# Шаг 2. Предобработка данных

## Обработка пропусков

```python
#заполним пропуска в стаже
data['days_employed'] = data['days_employed'].fillna(0)
data.info()
```

```python
# делаю список из уникальных значений типа занятости
# считаю в цикле для каждого типа среднюю и подставляю её в пропущенной значение этого типа
unique_income_types = data['income_type'].unique()
for in_type in unique_income_types:
    median = data[data['income_type'] == in_type]['total_income'].median()
#    data[data['income_type'] == in_type]['total_income'].fillna(mean, inplace = True)       
#    вот так не сработало, а как ниже, сработало
    data.loc[data['income_type'] == in_type, 'total_income'] = data.loc[data['income_type'] == in_type, 'total_income'].fillna(median)
data.info()
```

## Вывод

Заменил пустые значения в стаже на 0, а в доходе пустые значения поменял на среднее по конкретному типу дохода. Теперь пропусков не осталось и можно двигаться дальше.

## Лемматизация

```python
# лишний раз посмотрим на цели и выделим основные
data['purpose'].unique()
```

Основные цели: жилье/недвижимость, автомобиль, свадьба, образование, ремонт
<br><i>На потом - будет любопытно, какое образование у тех, кто хочет взять кредит на образование</i>

```python
#леммитизировал все цели
m = Mystem()
#purpose = m.lemmatize(data.loc[1,'purpose'])
lems = m.lemmatize(data['purpose'].to_string())
```

```python
# посчитаем количество обращений
purposes_counter = Counter(lems)
purposes_counter
```

```python
#выглядит не очень, но в целом видно ранее выделенные цели, попробуем оформить репрезентативнее
# делаю список из целей, по которому буду обращаться к словарю с количеством лемм, полученному выше
key_purpose = ['жилье', 'недвижимость', 'автомобиль', 'свадьба','образование','ремонт']
purpose_table =[]
# переменная для суммирования количества целей, для проверки того, что не упустил ни одной леммы
check_sum = 0
print('Цель кредита, кол-во\n------------')
for key in key_purpose:
    print(key, purposes_counter.get(key))
    a = key, purposes_counter.get(key)
    check_sum += purposes_counter.get(key)
    purpose_table.append(a)
print('------------\nКол-во строк в таблице:', data.shape[0])
print('Количество целей:',check_sum)
```

## Вывод

Цели в порядке убывания популярности - недвижимость (в дальнейшем объединю с жильем), автомобиль, образование, свадьба и ремонт.

Решил сверить количество целей и размер таблицы. Целей нассчиталось больше чем строк в таблице.
Понял, что так произошло из-за ремонта, так как он встречается в паре с жильем, если его исключить, то цифры сходятся.
Если не верите, то уберите из key_purpose 'ремонт' либо отнемите от 21454-607 и проверьте сами :)  
В дальнейшем, при категоризации, в процессе добавления столбца с целью буду делать таким образом, чтобы ремонт победил жилье.

## Категоризация данных
Добавлю столбец с категорией цели:
- недвижимость
- автомобиль
- свадьба 
- образование
- ремонт

Добавлю столбец с типом категорией дохода:
1. rich       
2. high       
3. medium+    
4. medium     
5. medium-    
6. low  

```python
# будет работать через apply, беру строку, из столбца c целью ищу леммы, если в леммах есть ключевое слово, 
# то возвращаю его и вставляю в новый столбец
def add_purpose(row):
    try:
        lemma = m.lemmatize(row['purpose'])
        if 'автомобиль' in lemma:
            return 'автомобиль'
        if 'свадьба' in lemma:
            return 'свадьба'
        if 'образование' in lemma:
            return 'образование'
        if 'жилье' in lemma:
            #ремонт встречается во фразе с жильем
            if 'ремонт' in lemma:
                return 'ремонт'    
            else:
                return 'недвижимость'
        if 'недвижимость' in lemma:
            return 'недвижимость'
    except:
        return 'ошибка кат-ии цели'
data['clean_purpose'] = data.apply(add_purpose, axis = 1)
```

```python
print('Проверяем наличие ошибок\n',data[data['clean_purpose'] == 'ошибка кат-ии цели'],'\n------------------\n------------------')
print('Проверяем что категории цели проставились\n',data.head(5))
```

```python
# лишний раз проверю что у меня ремонт, а не жилье/недвижимость, все ок
data[data["clean_purpose"] == 'ремонт'].head(5)
```

```python
# при помощи describe определю границы для категоризации дохода
# буду использовать квантили, таким образом можно гарантировать, что группы примерно равнозначны по объему
# чтобы в дальнейшем произвоидлся расчет нужно поменять формат на числовой
income_description = data['total_income'].describe(percentiles=[.16,.32,.48,.64,.82]).apply("{:.2f}".format)
income_description = pd.to_numeric(income_description)
income_description
```

```python
# пойдем по той же схеме, что и с целями
def add_income_category(row):
    try:
        income = row['total_income']
        if income >= income_description[-2]:
            return '1. rich'
        elif income_description[-3] <= income < income_description[-2]:
            return '2. high+'
        elif income_description[-5] <= income < income_description[-3]:
            return '3. high-'
        elif income_description[-6] <= income <  income_description[-5]:
            return '4. medium+'
        elif income_description[-7] <= income < income_description[-6]:
            return '5. medium-'
        elif income < income_description[-7]:
            return '6. low'
    except:
        return 'ошибка кат-ии дохода'
    
data['income_category'] = data.apply(add_income_category, axis = 1)
```

```python
print('Проверяем наличие ошибок\n',data[data['income_category'] == 'ошибка кат-ии дохода'],'\n------------------\n------------------')
print('Проверяем что категории цели проставились\n',data.head(5))
```

```python
# посмотрим долю каждого типа дохода от общего кол-ва клиентов
# использовал для калибровки значений, чтобы сильно большого разброса в середине не было
data.groupby('income_category')['income_category'].count()/data.shape[0]
```

## Вывод

При помощи леммитизации выделил из цели кредита ключевое слово и добавил его в новы столбец. Все прошло без ошибок. 
Основной нюанс был с ремонтом, который мог быть спутан с жильем и объединением жилья и недвижимости в одну категорию.

Выделил 6 категорий дохода, по моей классификации в банке есть небольшое количество богатых, еще меньшее количество бедных
Так же решил выделить три подкатегории для средного дохода. Интересно что получится при просмотре зависимости возврата кредита

## Исправление количества детей и выделение категорий по количеству детей

```python
# делаю срез по строчкам со значение ребенка -1 и заменяю на 1, затем проверяю
data.loc[data['children'] == -1,'children'] = 1
data[data['children'] == -1]
```

```python
# тоже самое проделываю для 20 детей
data.loc[data['children'] == 20,'children'] = 0
data[data['children'] == 20]
```

```python
# пойдем по той же схеме, что и с целями и доходом
def add_child_category(row):
    try:
        children = row['children']
        if children == 0:
            return 'нет детей'
        if 0 < children < 3:
            return '1-2 ребенка'
        if children >= 3:
            return 'многодетный'
    except:
        return 'ошибка кат-ии детей'
    
data['children_category'] = data.apply(add_child_category, axis = 1)
```

```python
# на этом шаге узнать про df.empty
print('Определение категорий прошло без ошибок?',data[data['income_category'] == 'ошибка кат-ии дохода'].empty,'\n------------------\n------------------')
print('Проверяем что категории цели проставились\n',data.head(5))
```

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
# строим сводную, так как наличие просрочки это "1", то суммируем и получаем количество клиентов с просрочками в разбивке по количеству детей
children_pivot = data.pivot_table(index = ['children_category','children'], values=['debt'], aggfunc=['sum','mean'])
children_pivot.columns = ['#', "share"]
format_d={'share':'{:.2%}'}
children_pivot.style.format(format_d)
```

На первый взгляд может показаться, что многодетные самые дисциплинированные, а без детей самые не дисциплинированные.
Но осознал, что смотреть в абсолюте - некорректно, посмотрим на доли. Плюс, понял что не было особого смысла выделять категории, так как значений детей всего 6. 

```python
# поделю сгруппированных по количеству детей клиентов с просрочкой на тех кто без просрочки 
children_share = ((data.loc[data['debt'] == 1,'children'].value_counts()) / (data['children'].value_counts()) * 100).fillna(0)
print('Кол-во детей | Доля клиентов в просрочке(%) с указанным количеством детей')
children_share.sort_values()
# пока шла проверка, узнал как можно такую же долю выводить в сводну без такого длинного кода
```

## Вывод

Таким образом, вышло как раз наоборот. Оказалось, что клиенты без детей платят лучше. Клиентов с 5б. слишком мало, чтобы считать их значимыми. 

- Есть ли зависимость между семейным положением и возвратом кредита в срок?

Для начала посмотрим сколько унас тех или иных клиентов с разным семейным положением

```python
data['family_status'].value_counts()
```

```python
# построим сводную, с учетом детей у клиента, заполним путые значения нулями
family_pivot = data.pivot_table(index = 'family_status', columns='children', values='debt', aggfunc=['sum','mean'], margins=True,fill_value=0)
family_pivot
```

```python
# визуально женатые намного больше попадают  в просрочку чем все остальные. Однако таких клиентов в принципе больше чем остальных.
# сделаем тоже упражнение что и для детей 
family_share = ((data.loc[data['debt'] == 1,'family_status'].value_counts()) / (data['family_status'].value_counts()) * 100)
family_share.sort_values()
```

## Вывод

Как и в прошлый раз, больше не значит хуже. Самые дисциплинированные клиенты - вдовы/вдовцы, разведеные тоже ведут себя не плохо, как и женатые. То есть те, кто был когда либо в браке или находятся в браке понимают, что такое ответственность.
В то время как наиболее фривольно себя чувствуют люди не женатые и в гражданском браке.

```python
# выведу сводную не только по доходу, но и в разрезе типа дохода
income_pivot = data.pivot_table(index = 'income_category', columns='income_type', values='debt', aggfunc='sum', margins=True, fill_value=0)
income_pivot
```

Интересно, что должников с medium+, больше чем с medium-, хотя козалось бы они состоятельнее. 
И на сколько помню, их доля была примерно равна. Уже традиционно выведу доли должников от общего количества.

```python
income_share = ((data.loc[data['debt'] == 1,'income_category'].value_counts()) / (data['income_category'].value_counts()) * 100)
income_share.sort_values()
```
