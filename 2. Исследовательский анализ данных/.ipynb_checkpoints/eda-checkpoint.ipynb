```python
# на будущее создам словарь и функцию для определения средней и медианы любого столбца для указанного города 
unique_localities = data['locality_name'].unique()
def locality_mean_median(locality, column):
    '''
    На вход подается локалити и столбец, по которому нужно рассчитать среднюю и медиану
    На выходе получаем список, первое число - средняя, второе - медиана
    '''
    locality_mean = round(data.loc[(data['locality_name'] == locality) & (data[column].notna()), column].mean(),ndigits=2)
    locality_median = round(data.loc[(data['locality_name'] == locality) & (data[column].notna()), column].median(),ndigits=2)
    stats = [locality_mean, locality_median]
    return stats
```

```python
ceiling_height_median_by_floor_dict = {}
for floor in floors_total_list:
    median = data.loc[data['floors_total'] == floor, 'ceiling_height'].median()
    ceiling_height_median_by_floor_dict[floor] = median
ceiling_height_median_by_floor_dict
```

```python
def ceiling_height_null_termination(row):
    try:
        if row['ceiling_height'] != row['ceiling_height']:
            floor = row['floors_total']
            median = ceiling_height_median_by_floor_dict.get(floor)
            return median
        else:
            same = row['floors_total']
            return same
    except:
        return 'Ошибка!'
```

```python
data['ceiling_height'] = data.apply(ceiling_height_null_termination, axis=1)
```

```python
#добавлю новый столбец, рассчитаю какой % от общей площади у других квартир занимает кухня
data.insert(13,'kitchen_ratio',data['kitchen_area']/data['total_area'])
```

Предположим, есть зависимость от количества комнат. 
Найдем среднее значение kitchen_ratio для каждого количества комнат и заполним пропуски исходя из него 

```python
# создам словарь с соответствием {количество комнат: средняя доля кухни от общей площади}
rooms_list = data['rooms'].unique()
share_dict = {}
for room in rooms_list: 
    share = data.loc[(data['rooms'] == room) & (data['kitchen_area'].notna()), 'kitchen_ratio'].mean()
    share_dict[room] = share
# проверка словаря
share_dict
```

```python
# напишем функцию, которую применим через apply к данным в таблице
# внутри функции проверяю заполнена площадь кухни или нет
# если заполнена, то расчитываю площадь кухни исходя из данных словаря и общей площади и возвращаю её
# если поле не пустое, то возвращаю текущее значение
def kitchen_area_calculation(row):
    if row['kitchen_area'] != row['kitchen_area'] :
        share_from_dict = share_dict.get(row['rooms'])
        kitchen_area = row['total_area'] * share_from_dict
        return kitchen_area
    else:
        same = row['kitchen_area']
        return same
```

```python
data['kitchen_area'] = data.apply(kitchen_area_calculation, axis=1)
```

```python
# заведем функция для определения числа корзин при помощи правила Стёрджеса
def bins_calculation(data,column):
    '''
    Функция для рассчета оптимального количества корзин
    для гистограммы по правилу Стёрджеса
    Для поиска значения укажите название df и колонки
    Для которой будет строиться гистограмма
    '''
    N = len(data[column].unique())
    bins = round(1 + math.log(N,2))
    return bins
```

```python
columns_to_research = ['total_area','last_price', 'rooms','ceiling_height'] 
# строю основу для размещения графиков в сабплоте 2х2
fig, axs = plt.subplots(2,2)
# координаты в размещения в сабплот, в цикле буду проверять значение counter
# чтобы в каждой итерации выбирались нужные координаты
axs_1=0
axs_2=0
counter = 0
for column in columns_to_research:
    q1=data[column].describe()[4]
    q3=data[column].describe()[6]
    iqr=q3-q1
    range_1 = q1 - iqr * 1.5
    range_2 = q3 + iqr * 3
    bins_n = bins_calculation(data,column)
    data[column].plot(kind='hist',
                      bins=bins_n, 
                      range=(range_1, range_2),
                      figsize=(20, 10), 
                      ax=axs[axs_1,axs_2], 
                      title='{} research'.format(column),
                     grid = True)    
    if counter == 0:
        axs_2 = 1
        counter += 1
    elif counter == 1:
        axs_1 = 1
        axs_2 = 0
        counter += 1
    elif counter == 2: 
        axs_2 = 1
```

```python
# определим выбросы для интересующих нас столбцов, соберем значение Q1  и верхнего "уса" в словарь
columns_to_cut = ['days_exposition','ceiling_height','rooms', 'last_price', 'total_area']
fig, axs = plt.subplots(2,3)
# координаты в размещения в сабплот, в цикле буду проверять значение counter
# чтобы в каждой итерации выбирались нужные координаты
axs_1=0
axs_2=0
counter = 0
box_plot_edges = {}
for column in columns_to_cut:
    q1=data[column].describe()[4]
    q3=data[column].describe()[6]
    iqr=q3-q1
    bottom_edge = int(q1)
    top_edge = int(q3 + 1.5 * iqr)
    data[column].plot(kind='box',
                      figsize=(20, 10), 
                      ax=axs[axs_1,axs_2], 
                      title='{} research\n {} top edge of boxplot'.format(column, top_edge ),
                     grid = True,
                     use_index = False)    
    box_plot_edges[column] = [bottom_edge]
#     for key in box_plot_edges:
#         if key in box_plot_edges:
    box_plot_edges[column].append(top_edge)
    if counter == 0:
        axs_2 = 1
        counter += 1
    elif counter == 1:
        axs_2 = 2
        counter += 1
    elif counter == 2: 
        axs_1 = 1
        axs_2 = 0
        counter += 1
    elif counter == 3:
        axs_1 = 1
        axs_2 = 1    
    elif counter == 4:
        axs_1 = 1
        axs_2 = 2  
        
```

- Самый дорогой квадратный метод из топ-10 населенных пунктов по количеству продавамой недвижмости- СПб
- Наличие аэропорта в населенном пункте увеличивает стоимость недвижимости, а лучшая, с точки зрения цены, удаленность от аэропорта это 30-40 км. 

Разработчикам необходимо передать, что наблюдаются проблемы с исходными данными. Один из вариантов решения - модернизация форм заполнения объявлений таким образом, чтобы у пользователя не было шанса ввести некорректное значение.
